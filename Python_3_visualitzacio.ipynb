```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np


# Dades d'exemple
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plotejar un gràfic
plt.plot(x, y, label='Sinus')
plt.legend()
plt.show()
```

```python
 #Exercici 3:
#Ploteja una gràfica de cos(x) sobre el mateix x i personalitza el color i títol.

# Dades d'exemple
x = np.linspace(0, 10, 100)
y = np.cos(x)

# Plotejar un gràfic
plt.plot(x, y, label='Cosinus')
plt.xlabel("X Value",  color = "purple")
plt.ylabel("Y Value", color = "purple")
plt.title("Cosinus graph", color = "brown")
plt.legend()
plt.show()
```
